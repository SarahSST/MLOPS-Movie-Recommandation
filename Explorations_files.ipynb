```python
movies = pd.read_csv('data/movies.zip',compression='zip', sep=',')

# Séparation titre et année
movies['year'] = movies['title'].str.extract('.*\((.*)\).*',expand = False)
movies['title'] = movies['title'].apply(lambda x: x[:-7])

# Séparation de la colonne genre
movies['genres'] = movies['genres'].str.split('|').values
movies = movies.explode(column=['genres'])
movies = movies.join(pd.get_dummies(movies['genres'], prefix='genre_'))
movies = movies.drop(['genres'], axis=1)
movies = movies.groupby(['movieId', 'title', 'year']).sum().reset_index()

print(movies.info())
movies.head(5)
```
